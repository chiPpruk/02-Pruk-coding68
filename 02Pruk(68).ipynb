# 17 ‡∏Å‡∏£‡∏Å‡∏é‡∏≤‡∏Ñ‡∏° 2568 ‡∏™‡∏≠‡∏ö‡∏Å‡∏•‡∏≤‡∏á‡∏†‡∏≤‡∏Ñ‡∏ó‡∏µ‡πà‡∏á‡πà‡∏≤‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î

```python
def ielts_to_cefr():
    while True:
        try:
            ielts_score = float(input('\n‡πÇ‡∏õ‡∏£‡∏î‡∏õ‡πâ‡∏≠‡∏ô ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô IELTS ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì (0.0 - 9.0): '))
            if 0.0 <= ielts_score <= 9.0:
                break
            else:
                print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô IELTS ‡∏ï‡πâ‡∏≠‡∏á‡∏≠‡∏¢‡∏π‡πà‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 0.0 ‡∏ñ‡∏∂‡∏á 9.0 ‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏•‡∏≠‡∏á‡πÉ‡∏´‡∏°‡πà‡∏≠‡∏µ‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á")
        except ValueError:
            print("‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏õ‡πâ‡∏≠‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô")

    if ielts_score >= 8.5:
        cefr_level = "C2 (Mastery)"
    elif ielts_score >= 7.0:
        cefr_level = "C1 (Advanced)"
    elif ielts_score >= 5.5:
        cefr_level = "B2 (Upper-Intermediate)"
    elif ielts_score >= 4.0:
        cefr_level = "B1 (Intermediate)"
    elif ielts_score >= 3.0:
        cefr_level = "A2 (Elementary)"
    elif ielts_score >= 1.0:
        cefr_level = "A1 (Beginner)"
    else:
        cefr_level = "‡∏ï‡πà‡∏≥‡∏Å‡∏ß‡πà‡∏≤ A1 (Pre-A1)"
    print(f"\n‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô IELTS ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì", ielts_score, "‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏£‡∏∞‡∏î‡∏±‡∏ö" , cefr_level)


def cefr_to_ielts():
    while True:
        cefr_level = input('\n‡πÇ‡∏õ‡∏£‡∏î‡∏õ‡πâ‡∏≠‡∏ô‡∏£‡∏∞‡∏î‡∏±‡∏ö CEFR ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì (A1, A2, B1, B2, C1, C2): ').upper()
        if cefr_level in ["A1", "A2", "B1", "B2", "C1", "C2"]:
            break
        else:
            print("‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏õ‡πâ‡∏≠‡∏ô‡∏£‡∏∞‡∏î‡∏±‡∏ö CEFR ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á (A1, A2, B1, B2, C1, C2)")

    if cefr_level == "C2":
        ielts_score = "8.5 - 9.0"
    elif cefr_level == "C1":
        ielts_score = "7.0 - 8.0"
    elif cefr_level == "B2":
        ielts_score = "5.5 - 6.5"
    elif cefr_level == "B1":
        ielts_score = "4.0 - 5.0"
    elif cefr_level == "A2":
        ielts_score = "3.0 - 3.5"
    elif cefr_level == "A1":
        ielts_score = "1.0 - 2.0"
    print(f"\n‡∏£‡∏∞‡∏î‡∏±‡∏ö CEFR ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì", cefr_level, "‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô IELTS ‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì", ielts_score)

while True:
    name = input("‡∏õ‡πâ‡∏≠‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì: ")
    print("‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ‡∏Ñ‡∏£‡∏±‡∏ö", name)
    print("\nWelcome to ‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô")

    while True:
        print("\n‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Å‡∏≤‡∏£‡πÅ‡∏õ‡∏•‡∏á:")
        print("1. IELTS to CEFR")
        print("2. CEFR to IELTS")
        choice = input("‡πÄ‡∏•‡∏∑‡∏≠‡∏Å (1/2): ")
        if choice in ['1', '2']:
            break
        else:
            print("‡πÇ‡∏õ‡∏£‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏Å 1 ‡∏´‡∏£‡∏∑‡∏≠ 2 ‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô")

    if choice == '1':
        ielts_to_cefr()
    elif choice == '2':
        cefr_to_ielts()

    another = input("\n‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏≠‡∏µ‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà? (yes/no): ")
    if another.lower() != 'yes':
        print("\n‡∏Ç‡∏≠‡∏ö‡∏Ñ‡∏∏‡∏ì‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏Ñ‡∏£‡∏±‡∏ö")
        print("\n‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏° By ‡πÄ‡∏•‡∏Ç‡∏ó‡∏µ‡πà 02 ‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏•‡πà‡∏ô: ‡∏û‡∏§‡∏Å‡∏©‡πå")
        print("Thank you kub")
        break
```
